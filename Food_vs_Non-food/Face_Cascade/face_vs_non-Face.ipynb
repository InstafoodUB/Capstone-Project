# Face vs Non-face 

Discriminació entre menjar i no menjar utilitzant detecció de cares (faceCascade).

```python
import cv2
import sys
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
from skimage.color import gray2rgb
import numpy as np
import copy
%matplotlib inline


imagePath = 'face_Nface/'
cascPath = '/home/eredxiii/anaconda2/pkgs/opencv3-3.1.0-py27_0/share/OpenCV/haarcascades/haarcascade_frontalface_default.xml'

face_images = [plt.imread(imagePath+'face'+str(i)+'.jpg') for i in range(1,6)]
no_face_images = [plt.imread(imagePath+'food'+str(i)+'.jpg') for i in range(1,6)]
images = face_images+no_face_images
images_copy =  copy.deepcopy(images)
```

```python
#show images
for i in range(10):
    plt.subplot('25'+str(i))
    plt.axis('off')
    plt.title('image'+str(i))
    plt.imshow(images[i])
```

```python
# Create the haar cascade

for i in range(10):
    faceCascade = cv2.CascadeClassifier(cascPath)
    gray = gray2rgb(images[i])
    
    # Detect faces in the image
    faces = faceCascade.detectMultiScale(
        gray,
        scaleFactor=1.25,
        minNeighbors=5,
        minSize=(30, 30)
    )

    print 'Found {0} faces in image '.format(len(faces))+str(i)+'!'

    # Draw a rectangle around the faces
    for (x, y, w, h) in faces:
        plt.figure(i)
        plt.title('image'+str(i))
        plt.imshow(cv2.rectangle(images_copy[i], (x, y), (x+w, y+h), (0, 255, 0), 2))
```
